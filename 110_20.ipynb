```python
#Create a clock using geometry and trigonometry

import random as ra
import matplotlib.pyplot as plt
import numpy as np
import math
from math import cos, sin

for i in range(0,10) :
  #Make the circle
  fig = plt.figure(figsize = (4,4))
  ax = fig.add_subplot(111)

  theta = np.linspace(0, 2*np.pi, 100)
  r = 3

  xs = r * np.cos(theta)
  ys = r * np.sin(theta)

  ax.plot(xs, ys, color = 'black')

  #Initialize hour 3 and minute 15 dash coordinates
  xh0 = 3
  xh1 = 2.5
  yh0 = 0
  yh1 = 0

  xm0 = 3
  xm1 = 2.8
  ym0 = 0
  ym1 = 0

  #Plot center point
  plt.scatter(0,0, color = 'black')

  #Plot dashes for each hour
  for i in range(0, 360, 30) :
    angle = i / 180 * math.pi
    x10 = xh0 * cos(angle) - yh0 * sin(angle)
    x11 = xh1 * cos(angle) - yh1 * sin(angle)
    y10 = xh0 * sin(angle) + yh0 * cos(angle)
    y11 = xh1 * sin(angle) + yh1 * cos(angle)

    ax.plot([x10,x11], [y10,y11], color = 'black')
    ax.axis('off')


  #Plot dashes for each minute
  for i in range(0, 360, 6) :
    angle = i / 180 * math.pi
    x10 = xm0 * cos(angle) - ym0 * sin(angle)
    x11 = xm1 * cos(angle) - ym1 * sin(angle)
    y10 = xm0 * sin(angle) + ym0 * cos(angle)
    y11 = xm1 * sin(angle) + ym1 * cos(angle)

    ax.plot([x10,x11], [y10,y11], color = 'black')
    ax.axis('off')

  #Make choices for hours/minutes and get associated angles
  hours = list(range(12,0,-1))
  hours_3 = hours[hours.index(3):] + hours[:hours.index(3)]

  minutes = list(range(59,-1,-1))
  minutes_15 = minutes[minutes.index(15):] + minutes[:minutes.index(15)]

  thetas_h = list(range(0,360,30))
  hour = ra.choice(hours_3)
  ind_h = hours_3.index(hour)
  theta_h = thetas_h[ind_h]

  thetas_m = list(range(0,360,6))
  min = ra.choice(minutes_15)
  ind_m = minutes_15.index(min)
  theta_m = thetas_m[ind_m]

  #Convert thetas to radians
  theta_h = theta_h * math.pi / 180
  theta_m = theta_m * math.pi / 180

  #Hour Hand
  hx = 1.2
  hy = 0
  hour_x = hx * cos(theta_h) - hy * sin(theta_h)
  hour_y = hx * sin(theta_h) + hy * cos(theta_h)

  plt.arrow(0,0, hour_x, hour_y, width = 0.075, color = 'black')

  #Minute Hand
  mx = 2.5
  my = 0
  min_x = mx * cos(theta_m) - my * sin(theta_m)
  min_y = mx * sin(theta_m) + my * cos(theta_m)

  plt.arrow(0,0, min_x, min_y, width = 0.075, color = 'black')

  #Choose which hand they are giving an answer for
  hands = ['hour', 'minute']
  hand = ra.choice(hands)
  print('Which {} is the {} hand pointing at?'.format(hand, hand))
  plt.show()
  if hand == 'hour' :
    print(hour)
    print('\n\n')
  else :
    print(min)
    print('\n\n')
```
