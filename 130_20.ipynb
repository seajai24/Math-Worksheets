```python
#Graphing images and money identification

import random as ra
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

dollar = Image.open('dollar.png')
quarter = Image.open('quarter.png')
dime = Image.open('dime.png')
nickel = Image.open('nickel.png')
penny = Image.open('penny.png')
half = Image.open('half.png')
five = Image.open('5dollar.png')
hundo = Image.open('100dollar.png')

#Width Ratios -- Dime: 0.75, Penny: 1, Nickel: 1.25, Quarter: 1.5, Half: 2.25, Bills: 3.5
coins = {'$1.00' : 3.5, '$0.25' : 1.5, '$0.10' : 0.75, '$0.05' : 1.25, '$0.01' : 1, '$0.50' : 2.25, '$5.00' : 3.5, '$100.00' : 3.5}

c_names = list(coins.keys())

print('Give the value for each.\n')

for j in range(0,6) :

  cs = ra.sample(c_names, k=3)
  
  fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (12,5), gridspec_kw={'width_ratios': [coins[cs[0]], coins[cs[1]], coins[cs[2]]]})

  for i in range(0,3) :
    if i == 0 :
      if cs[i] == '$1.00' :
        ax1.imshow(dollar)
      elif cs[i] == '$0.25' :
        ax1.imshow(quarter)
      elif cs[i] == '$0.10' :
        ax1.imshow(dime)
      elif cs[i] == '$0.05' :
        ax1.imshow(nickel)
      elif cs[i] == '$0.01' :
        ax1.imshow(penny)
      elif cs[i] == '$0.50' :
        ax1.imshow(half)
      elif cs[i] == '$5.00' :
        ax1.imshow(five)
      else :
        ax1.imshow(hundo)
    elif i == 1 :
      if cs[i] == '$1.00' :
        ax2.imshow(dollar)
      elif cs[i] == '$0.25' :
        ax2.imshow(quarter)
      elif cs[i] == '$0.10' :
        ax2.imshow(dime)
      elif cs[i] == '$0.05' :
        ax2.imshow(nickel)
      elif cs[i] == '$0.01' :
        ax2.imshow(penny)
      elif cs[i] == '$0.50' :
        ax2.imshow(half)
      elif cs[i] == '$5.00' :
        ax2.imshow(five)
      else :
        ax2.imshow(hundo)
    else :
      if cs[i] == '$1.00' :
        ax3.imshow(dollar)
      elif cs[i] == '$0.25' :
        ax3.imshow(quarter)
      elif cs[i] == '$0.10' :
        ax3.imshow(dime)
      elif cs[i] == '$0.05' :
        ax3.imshow(nickel)
      elif cs[i] == '$0.01' :
        ax3.imshow(penny)
      elif cs[i] == '$0.50' :
        ax3.imshow(half)
      elif cs[i] == '$5.00' :
        ax3.imshow(five)
      else :
        ax3.imshow(hundo)

  ax1.axis('off')
  ax2.axis('off')
  ax3.axis('off')

  plt.show()

  print('a) {} \t\t b) {} \t\t c) {}\n\n'.format(cs[0], cs[1], cs[2]))
```
